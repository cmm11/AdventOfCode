```python
fname = "/Users/carmenmerx/Documents/GitHub/AdventOfCode/05AoCInput.txt"
with open(fname) as f:
    content = f.readlines()
content = [x.strip() for x in content] 
```

```python
content
```

```python
input_string = content[0]
```

```python
import string
lc = string.ascii_lowercase
uc = string.ascii_uppercase


luc = ["".join([a,b]) for a, b in zip(lc, uc)]
ulc = ["".join([b,a]) for a, b in zip(lc, uc)]

print(luc)
print(ulc)
```

```python
def collapse_polymer(content):

    length = len(content)
    new_length = 0

    while (length != new_length):

        length = new_length

        for i in luc:

            content = content.replace(i, "")

        for i in ulc:

            content = content.replace(i, "")


        new_length = len(content)

    return(len(content))

```

```python
letters = {}
for i, letter in enumerate(lc):
    
    content = input_string
    content = content.replace(letter, "")
    content = content.replace(uc[i], "")
    length = collapse_polymer(content)
    letters[letter] = length
    
```

```python
letters
```

```python
minimum = 10000000

for letter in letters:
    if letters[letter] < minimum:
        minimum = letters[letter]
```

```python
import numpy as np
import pandas as pd
import datetime
from dateutil.relativedelta import relativedelta

df = pd.DataFrame(columns = ["Timestamp", "Guard", "Action"])

for i in content:
    guard = np.nan
    split = i.split("]")
    action = split[1].strip()
    if (len(action.split(" ")) == 4):
        guard = action.split(" ")[1]
        action = " ".join([action.split(" ")[2], action.split(" ")[3]])
    timestamp = split[0].split("[")[1].strip()
    timestamp = datetime.datetime.strptime(timestamp, "%Y-%m-%d  %H:%M")
    timestamp = timestamp + relativedelta(years=500)
    timestamp = pd.to_datetime(timestamp)


    df = df.append({"Timestamp" : timestamp, "Guard": guard, "Action" : action}, ignore_index=True)

df = df.sort_values(by = "Timestamp")
df = df.fillna(method = "ffill")
```

```python
guards = {}
minutes_guard = {}

for i in range(0, df.shape[0]):
    
    minutes = np.zeros(60)

    if (df.iloc[i]["Action"] == "falls asleep"):
        start = df.iloc[i]["Timestamp"].minute
        end = df.iloc[i+1]["Timestamp"].minute
        diff = end - start
        guard = df.iloc[i]["Guard"]
        
        for j in range(start, end):
            minutes[j] += 1
        if guard not in guards:
            guards[guard] = diff
        else:
            guards[guard] += diff
            
        if guard not in minutes_guard:
            minutes_guard[guard] = minutes
        else: 
            minutes_guard[guard] = [x + y for x, y in zip(minutes_guard[guard], minutes)]

```

```python
maximum = 0
max_guard = ""

for guard in guards:
    if guards[guard] > maximum:
        maximum = guards[guard]
        max_guard = guard
print(maximum)
print(max_guard)
```

```python
maximum_min = 0
min_index = 99
max_guard = ""

for guard in minutes_guard:
    if max(minutes_guard[guard]) > maximum_min:
        maximum_min = max(minutes_guard[guard])
        min_index = minutes_guard[guard].index(maximum_min)
        max_guard = guard
print(maximum_min)
print(min_index)
print(max_guard)
```
